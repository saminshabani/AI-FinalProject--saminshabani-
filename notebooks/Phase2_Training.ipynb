```python
import cv2

def predict_single_image(image_path, model, class_names):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None: return
    
    # Ù¾ÛŒØ´â€ŒÙ¾Ø±Ø¯Ø§Ø²Ø´ Ø¯Ù‚ÛŒÙ‚ Ù…Ø´Ø§Ø¨Ù‡ Ø¢Ù…ÙˆØ²Ø´
    img_resized = cv2.resize(img, (224, 224))
    img_input = img_resized.astype('float32') / 255.0
    img_rgb = np.stack([img_input]*3, axis=-1)
    img_batch = np.expand_dims(img_rgb, axis=0)
    
    preds = model.predict(img_batch, verbose=0)
    pred_idx = np.argmax(preds)
    actual_class = os.path.basename(os.path.dirname(os.path.dirname(image_path)))
    
    color = 'green' if class_names[pred_idx] == actual_class else 'red'
    plt.imshow(img_resized, cmap='gray')
    plt.title(f"Actual: {actual_class}\nPred: {class_names[pred_idx]}", color=color)
    plt.axis('off')

# ØªØ³Øª Ø±ÙˆÛŒ Ú†Ù†Ø¯ Ù†Ù…ÙˆÙ†Ù‡
samples = [
    '../data/raw/COVID/images/COVID-1.png',
    '../data/raw/Lung_Opacity/images/Lung_Opacity-1008.png'
]

plt.figure(figsize=(15, 5))
for i, path in enumerate(samples):
    if os.path.exists(path):
        plt.subplot(1, 3, i+1)
        predict_single_image(path, model, class_names)
plt.show()
```
